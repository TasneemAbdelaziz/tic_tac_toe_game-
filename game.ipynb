```python
import numpy as np
board = np.zeros((3,3))
# print(board)

#display game board
def GameBoard(board):

    #game rule
    chars = {1:'X', -1:'O',0:' _ '}

    #row by row
    for x in board:
        #element by element
        for y in x:
            #access values in chars dic by using key y
            # 0 -> _
            # 1 -> X (player)
            # -1 -> O (AI)
            ch = chars[y]
            print(ch,end=' ')
        print()


#clear board 
def ClearBoard(board):
        board[:] = 0

#check win
def WinningPlayer(board,player):
    conditions=[
        #horizontal 
        [board[0][0],board[0][1],board[0][2]],
        [board[1][0],board[1][1],board[1][2]],
        [board[2][0],board[2][1],board[2][2]],
        #vertical
        [board[0][0],board[1][0],board[2][0]],
        [board[0][1],board[1][1],board[2][1]],
        [board[0][2],board[1][2],board[2][2]],
        #cross
        [board[0][0],board[1][1],board[2][2]],
        [board[0][2],board[1][1],board[2][0]]
    ]

#if there is [1,1,1] or [-1,-1,-1] in the same inner array like [b[0][0],b[0][1],b[0][2]]
    if [player,player,player] in conditions:
        return True
    return False

#----------------------------------------------------------
#check who win Computer (-1) or Player (1)
def gameWon(board):
    return WinningPlayer(board,1) or WinningPlayer(board,-1)
#------------------------------------------------------------

#print game result
def printResult(board):
    if WinningPlayer(board,1):
        print("Congratulations! You won the game.")
    elif WinningPlayer(board,-1):
        print("Sorry. The AI has won this time.")
    else:
        print("It's a draw.")


#check blank positions
def blanks(board):
    #array for to store in it blank position
    blank =[]
    # x -> index of each row 0 - 1 - 2 - 3
    # row -> element of this row [0,0,0]
    for x,row in enumerate(board):
        # y -> index of each element in row 
        # col -> value of element in col
        for y,col in enumerate(row):
            #x,y -> (0 , 0) - (0 , 1) - (0 , 2)
            if board[x][y] == 0:
                blank.append([x,y])
    return blank

#Is FULL
def boardFull(board):
    if len(blanks(board)) == 0:
        return True
    return False

#Set player move on the board for AI(-1) or human(1)
def setMove(board,x,y,player):
    board[x][y] = player

#game state
def evaluate(board):
    if WinningPlayer(board,1):
        return 1
    elif WinningPlayer(board,-1):
        return -1
    else:
        return 0 

#------------------------------
def heuristic(board):
    return evaluate(board)
#---------------------------

GameBoard(board)
ClearBoard(board)
WinningPlayer(board,1)
blanks(board)
```
